```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## 기본 그래프(선 그래프 그리기)



```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], label='line') 
plt.title('title')          # 표 제목
plt.xlabel('x-axis')        # x축 라벨
plt.ylabel('y-axis')        # y축 라벨
plt.legend()                # 범례 표시
plt.grid()                  # 격자 표시
plt.show()
```

```python
# 선, 마커, 색상 지정하기

plt.plot([1, 2, 3], [4, 4, 4], 'bo-')
plt.plot([1, 2, 3], [3, 3, 3], 'rv--')
plt.plot([1, 2, 3], [2, 2, 2], 'g^-.')
plt.plot([1, 2, 3], [1, 1, 1], 'ys:')
plt.axis([0.6, 3.4, 0.6, 4.5])            # 표현할 축의 범위 나타내기 순서대로 [xmin, xmax, ymin, ymax]
plt.show()
```

```python
# 그래프 한번에 여러개 그리기

x1 = np.arange(0, 11)
y1 = x1 ** 2
x2 = np.arange(0, 11)
y2 = (x2 ** 2) * 2
plt.title('title')          # 표 제목
plt.xlabel('x-axis')        # x축 라벨
plt.ylabel('y-axis')        # y축 라벨
plt.plot(x1, y1, label='y = x^2')
plt.plot(x2, y2, label='y = 2x^2')      # plt.plot(x1, y1, x2, y2) 로 한 줄에 한번에 적어도 같은 결과
plt.legend()                # 범례
plt.show()
```

```python
# 그래프 창을 여러 영역으로 나누고 싶을 때

x1 = np.arange(0, 50)
y1 = x1 * 2
x2 = np.arange(0, 50)
y2 = x2 * 4
plt.subplot(1, 2, 1)
plt.plot(x1, y1)
plt.subplot(1, 2, 2)
plt.plot(x2, y2)
plt.show()
```

```python
# 그래프의 일부분만 표시하고 싶을 때

x2 = np.arange(0, 50)
y2 = x2 * 4
plt.plot(x2, y2)
plt.xlim(20, 30)
plt.ylim(70, 150)
plt.show()
```

## 산점도(scatter)

`plt.scatter(x, y, [s=size_n, c=color, marker, alpha])`

x = x 값 리스트  

y = y 값 리스트

size = 각 점들의 크기

marker = 점의 모양

alpha = 투명도

```python
height = [158, 173, 174, 154, 177, 188, 183, 168, 185, 156, 160, 169, 167, 154, 163]
weight = [52, 66, 66, 49, 69, 79, 75, 61, 76, 51, 54, 62, 61, 49, 55]
plt.scatter(height, weight)
plt.xlabel('height')
plt.ylabel('weight')
plt.show()
```

```python
# 점의 색깔이나 크기 정하기
# 데이터가 3차원 혹은 4차원인 경우 점의 크기나 색깔을 이용하여 시각화 

np.random.seed(30)

n = 50
x = np.random.rand(n)
y = np.random.rand(n)
area = 500 * np.random.rand(n)
colors = np.random.rand(n)

plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.colorbar()        # 컬러바 표시하기
plt.show()
```

## 막대 그래프



```python
x = np.arange(3)
years = ['2018', '2019', '2020']
values = [100, 400, 900]

plt.subplot(2,1,1)
plt.bar(x, values)
plt.xticks(x, years)    # x축에 표시될 문자열 지정 
plt.subplot(2,1,2)
plt.barh(x, values)     # 수평 막대 그래프
plt.yticks(x, years)

plt.show()
```

## 파이 차트(pie chart)



```python
ratio = [64, 46, 18, 12]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']


plt.pie(ratio, labels=labels, autopct='%.1f%%')
plt.show()
```

```python
explode = [0, 0, 0.1, 0]

plt.pie(ratio, labels=labels, autopct='%.1f%%', \
        startangle=90, counterclock=False, explode=explode)
plt.show()
```

## 실습하기

캐글의 Fish market 데이터를 활용해서 시각화 연습해보자  
[Fish market](https://www.kaggle.com/datasets/aungpyaeap/fish-market)  
Length1 = vertical length  
Length2 = diagonal length  
Length3 = cross length

```python
from collections import Counter

df = pd.read_csv('/content/sample_data/Fish.csv')
df.head()
```

```python
species = df['Species']
names = []
counts = []
for key,values in Counter(species).most_common():
  names.append(key)
  counts.append(values)
print(names, counts)
```

```python
plt.pie(counts, labels=names, autopct='%.1f%%', counterclock=False, startangle=90)
plt.show()
```

```python
for i in range(len(species)):
  if species[i] != 'Bream':
    print(i)
    break

for i in range(len(species)):
  if species[i] == 'Perch':
    print(i)
    break
```

```python
bream_data = df.iloc[0:35]
perch_data = df.iloc[72:128]
perch_data.tail()
```

```python
# 스캐터 그리기

bream_weight = bream_data['Weight']
bream_length = bream_data['Length1']
perch_weight = perch_data['Weight']
perch_length = perch_data['Length1']

plt.scatter(bream_length, bream_weight, label='bream')
plt.scatter(perch_length, perch_weight, label='perch')
plt.xlabel('length')
plt.ylabel('weight')
plt.legend()
plt.show()
```

```python
# 히스토그램 그리기

ax1 = plt.subplot(1,2,1)
plt.hist(bream_weight, label='bream', color='#1a9988')
plt.legend()
ax2 = plt.subplot(1,2,2, sharex=ax1, sharey=ax1)
plt.hist(perch_weight, label='perch', color='#eb5600')
plt.legend()
plt.show()
```

```python
bream_mean = [bream_data['Length1'].mean(), bream_data['Length2'].mean(), bream_data['Length3'].mean()]
perch_mean = [perch_data['Length1'].mean(), perch_data['Length2'].mean(), bream_data['Length3'].mean()]

x = np.arange(3)
labels = ['Length1', 'Length2', 'Length3']

plt.bar(x, bream_mean, width=0.3, align='edge', label='bream', color='#1a9988')
plt.bar(x, perch_mean, width=-0.3, align='edge', label='perch', color='#eb5600')
plt.xticks(x, labels)

plt.legend()
plt.show()
```
